## Drzewo Trie

```python
class Trie:
    def __init__ (self, text=None):
        self.root = Node ()
        if text != None:
            self.make_trie(text)
        
    def make_trie (self, text):
        for i in range (len(text)):
            node = self.root
            for k, letter in enumerate(text [i:]):
                for nd in node.childs:
                    if nd.value == letter:
                        node = nd
                        break
                else:
                    self.graft (node, text[i+k:])
                    break
    
    def graft (self, head, text):
        for letter in text:
            new = Node()
            new.value = letter
            new.parent = head
            head.childs.append(new)
            head = new            
        
    def print_trie (self, node = None):
        if node == None:
            node = self.root
        if not node.childs:
            print (node.value, end=" ")
            return
        print (node.value, end="")
        print ("-> {", end=" ")
        for child in node.childs:
            self.print_trie(child)
        print ("}", end=" ")
```

Wykorzystam algorytm KMP z poprzednich zajęć. Przeprowadzę testy poszukiwań wzorca w tekście za pomocą algorytmu KMP oraz z wykorzystaniem drzew Trie i sufiksów. Funkcja `make_result` służy do policzenia liczby oraz długości wszystkich ścieżek w drzewie w którym korzeniem jest wierzchołek zawierający ostatni znak wzorca, przekształca ona te liczby aby stworzyć tablice z wynikami taką samą jaką tworzy algorytm KMP.

### Algorytm KMP

```python
def prefix_function (pattern):
    pi = [0]
    k = 0
    for q in range (1, len(pattern)):
        while k > 0 and pattern [k] != pattern [q]:
            k = pi[k-1]
        if pattern[k] == pattern[q]:
            k += 1
        pi.append(k)
    return pi

def kmp_string_matching (text, pattern):
    pi = prefix_function (pattern)
    result = []
    q = 0
    for i in range(0, len(text)):
        while q > 0 and pattern [q] != text[i]:
            q = pi[q-1]
        if pattern[q] == text [i]:
            q += 1
        if q == len(pattern):
            result.append (i + 1 - q)
            q = pi[q-1]
    return result
```

### Wyszukiwanie w drzewie Trie

```python
def make_result (node, res, n, l):
    t = len(node.value)
    if t>1:
        t = node.value[1]-node.value[0]
    l += t
    if not node.childs:
        res.append(n - l)
        return
    for nd in node.childs:
        make_result (nd, res, n, l)

def find_trie (text, pattern):
    trie = Trie (text)
    node = trie.root
    for letter in pattern:
        for nd in node.childs:
            if nd.value == letter:
                node = nd
                break
        else:
            return []
    res = []
    make_result (node, res, len(text), len(pattern)-1)
    return sorted(res)
```

### Wyszukiwanie w drzewie sufiksów

```python
def find_stree (text, pattern):
    stree = SuffixTree (text)
    node = stree.root
    k = 0
    while k < len(pattern):
        for nd in node.childs:
            if text[nd.value[0]] == pattern[k]:
                t = min (nd.value[1]-nd.value[0], len(pattern) - k)
                if pattern[k:t+k] == text[nd.value[0]:nd.value[0]+t]:
                    node = nd
                    k += (nd.value[1]-nd.value[0])
                    break
                else:
                    return []
        else:
            return []
    res = []
    make_result (node, res, len(text), k - (node.value[1]-node.value[0]))
    return sorted(res)
```

### Wyszukiwanie w drzewie sufiksów tworzonym algorytmem Ukkonena

```python
def find_utree (text, pattern):
    stree = UkkonenTree (text)
    node = stree.root
    k = 0
    while k < len(pattern):
        for nd in node.childs:
            if text[nd.value[0]] == pattern[k]:
                t = min (nd.value[1]-nd.value[0], len(pattern) - k)
                if pattern[k:t+k] == text[nd.value[0]:nd.value[0]+t]:
                    node = nd
                    k += (nd.value[1]-nd.value[0])
                    break
                else:
                    return []
        else:
            return []
    res = []
    make_result (node, res, len(text), k - (node.value[1]-node.value[0]))
    return sorted(res)
```
