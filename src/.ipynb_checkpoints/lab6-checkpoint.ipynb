## Wyszukiwanie wzorca dwuwymiarowego

KMP do wyszukiwania ostatecznego wzorca w pionie.

```python
def prefix_function (pattern):
    pi = [0]
    k = 0
    for q in range (1, len(pattern)):
        while k > 0 and pattern [k] != pattern [q]:
            k = pi[k-1]
        if pattern[k] == pattern[q]:
            k += 1
        pi.append(k)
    return pi

def kmp_string_matching (text, pattern, pi = None):
    if pi == None: pi = prefix_function (pattern)
    result = []
    q = 0
    for i in range(0, len(text)):
        while q > 0 and pattern [q] != text[i]:
            q = pi[q-1]
        if pattern[q] == text [i]:
            q += 1
        if q == len(pattern):
            result.append (i + 1 - q)
            q = pi[q-1]
    return result
```

Struktura węzła i drzewa Trie z linkami.

```python
class Node:
    def __init__ (self):
        self.value = None
        self.parent = self
        self.children = {}
        self.link = None
        self.end = 0
```

```python
class Trie:
    def __init__ (self, matrix):
        self.root = Node ()
        self.make_trie(matrix)
        
    def make_trie (self, matrix):
        for line in matrix:
            actual = self.root
            for k, el in enumerate(line):
                node = actual.children.get(el, None)
                if node != None:
                    actual = node
                else:
                    self.graft (actual, line[k:])
                    break
    
    def graft (self, head, arr):
        for el in arr:
            new = Node()
            new.value = el
            new.parent = head
            head.children[el] = new
            head = new
            
    def make_links (self):
        queue = deque()
        for child in self.root.children.values():
            child.link = self.root
            for ch in child.children.values():
                queue.append(ch)
        while len(queue) > 0:
            node = queue.popleft()
            el = node.value
            act = node.parent.link
            while True:
                nd = act.children.get(el, None)
                if nd != None:
                    act = nd
                    break
                elif act.link == None:
                    break
                else:
                    act = act.link
            node.link = act
            for child in node.children.values():
                queue.append(child)                    
        
    def print_trie (self, node = None):
        if node == None:
            node = self.root
        if not node.children:
            print (node.value, end=" ")
            return
        print (node.value, end="")
        print ("-> {", end=" ")
        for child in node.children.values():
            self.print_trie(child)
        print ("}", end=" ")
```

Wyszukiwanie wzorca dwuwymiarowego w macierzy. Najpierw tworzy drzewo Trie i za pomocą tego automatu wyszukuje położenie wszystkich lini. Następnie za pomocą kmp szukam w kolumnach położenie nowo utworzonego wzorca powstałego poprzez zapisanie każdej lini jako jednej liczby. Funkcja zwraca listę krotek, gdzie każda krotka to prawy dolny róg wzorca w macierzy.

```python
def search_2D_pattern (pattern, matrix):
    trie = Trie(pattern)
    trie.make_links ()
    pattern_1D = []
    it = 1
    for line in pattern:
        node = trie.root
        for el in line:
            node = node.children[el]
        if node.end:
            pattern_1D.append(node.end)
        else:
            node.end = it
            pattern_1D.append(it)
            it += 1
    res_search = np.zeros_like (matrix, dtype=np.int64)
    for i, line in enumerate(matrix):
        node = trie.root
        for j, el in enumerate(line):
            new = node.children.get(el, None)
            while new == None:
                if node.link == None:
                    new = node
                    break
                node = node.link
                new = node.children.get(el, None)
            node = new
            res_search[i, j] = node.end
    new_matrix = res_search.T
    pi = prefix_function (pattern_1D)
    l = len (pattern_1D)
    result = []
    for i, line in enumerate(new_matrix):
        res = kmp_string_matching(line, pattern_1D, pi)
        for j in res:
            result.append((j+l-1, i))
    return result       
```

## Ta sama litera jedna pod drugą

W celu wykonania tego zadania napiszę funkcję podobną do wyszukiwania wzorca dwuwymiarowego jednakże uproszczoną bo wzorzec dwóch liter pod sobą jest w pewnym sensie jednowymiarowy oraz wypiszę znalezienia dla każdej litery osobno. Zwracana pozycja wzorca to położenie dolnej litery.

```python
from string import ascii_letters

def search_letter_under_letter (matrix):
    pattern = np.array(list(ascii_letters)).reshape(-1, 1)
    pattern = np.hstack((pattern, pattern))
    trie = Trie(pattern)
    trie.make_links ()
    pattern_1D = []
    it = 1
    for line in pattern:
        node = trie.root
        for el in line:
            node = node.children[el]
        if node.end:
            pattern_1D.append(node.end)
        else:
            node.end = it
            pattern_1D.append(it)
            it += 1
    matrix = matrix.T
    res_search = np.zeros_like (matrix, dtype=np.int64)
    for i, line in enumerate(matrix):
        node = trie.root
        for j, el in enumerate(line):
            new = node.children.get(el, None)
            while new == None:
                if node.link == None:
                    new = node
                    break
                node = node.link
                new = node.children.get(el, None)
            node = new
            res_search[i, j] = node.end
    new_matrix = res_search.T
    result = {}
    for i, letter in enumerate(pattern[:,0]):
        result[letter] = np.argwhere(new_matrix == (i+1))
    return result       
```

```python
result = search_letter_under_letter (matrix_text)
for letter, positions in result.items():
    print ("Dla litery", letter)
    print (positions if len(positions) > 0 else "nie znaleziono")
```

## Pomiar czasu

Najpierw rozdzieliłem funkcje `search_2D_pattern` na swie części na których będę wykonywał pomiary czasu. Pierwsze sprawdziłem jak wielkość wzorca wpływa na czas budowania automatu oraz szukania wzorca.

```python
def build_automat (pattern):
    trie = Trie(pattern)
    trie.make_links ()
    pattern_1D = []
    it = 1
    for line in pattern:
        node = trie.root
        for el in line:
            node = node.children[el]
        if node.end:
            pattern_1D.append(node.end)
        else:
            node.end = it
            pattern_1D.append(it)
            it += 1
    return trie, pattern_1D

def search_pattern (trie, pattern_1D, matrix):
    res_search = np.zeros_like (matrix, dtype=np.int64)
    for i, line in enumerate(matrix):
        node = trie.root
        for j, el in enumerate(line):
            new = node.children.get(el, None)
            while new == None:
                if node.link == None:
                    new = node
                    break
                node = node.link
                new = node.children.get(el, None)
            node = new
            res_search[i, j] = node.end
    new_matrix = res_search.T
    pi = prefix_function (pattern_1D)
    l = len (pattern_1D)
    result = []
    for i, line in enumerate(new_matrix):
        res = kmp_string_matching(line, pattern_1D, pi)
        for j in res:
            result.append((i, j+l-1))
    return result
```

```python
OX = []
OB = []
OS = []
for size in range (10, 501, 20):
    pattern = matrix_png[:size, :size]
    OX.append(size)
    start = time()
    result = build_automat (pattern)
    end = time()
    OB.append(end-start)
    start = time()
    result = search_pattern (*result, matrix_png)
    end = time()
    OS.append(end-start)
OX = np.array(OX)
OB = np.array(OB)
OS = np.array(OS)
plt.figure(figsize=(20,30))
plt.subplot(311, title="Zależność czasu budowania od wielkości wzorca")
plt.plot (OX, OB)
plt.subplot(312, title="Zależność czasu szukania od wielkości wzorca")
plt.plot (OX, OS)
plt.subplot(313, title="Zależność czasu budowania + szukania od wielkości wzorca")
plt.plot (OX, OB + OS)
plt.show()
```

Im większy wzorzec tym dłużej trwa budowanie drzewa Trie. Jest to logiczne ponieważ zwiększamy drzewo Trie. Możemy zaobserwować że wykres przypomina oczekiwaną złożoność nlogn. Jednakże dla dużych wzorców zyskujemy na czasie wyszukiwania. Jest to spowodowane dużym dopasowaniem a więc nie ma potrzeby korzystanmia z linków oraz macierz do KMP ma dużo zer. Wykres sumaryczny ma minimum po środku. Oznacza to że najszybciej działa algorytm dla średniej wielkości wzorców w porównaniu do macierz.

```python
OX = []
OY = []
l = len (matrix_png) // 8
for i in range (1, 9):
    new_matrix = matrix_png[:(i*l)]
    OX.append(100 * i / 8)
    start = time ()
    _ = search_pattern (*result, new_matrix)
    end = time ()
    OY.append(end - start)
    
plt.figure(figsize=(20,10))
plt.plot(OX, OY)
plt.show()
```

Zgodnie z oczekiwaniami uzyskujemy zależność liniową od  wielkości obszaru. Jest to oczekiwana złożoność algorytmu.

Na koniec zbadam wpływ na czas szukania w całym obszarze a szukania w kawałkach. Dla poprawy wyników test przeprowadzę 4 razy i podam wynik sumaryczny.

```python
OX = []
OY = []
l = len (matrix_png)
for i in [1, 2, 4, 6, 8, 12, 16]:
    sum_time = 0
    step = l // i
    OX.append(i)
    for _ in range (4):
        for j in range (i):
            new_matrix = matrix_png[j*step:(j+1)*step]
            start = time ()
            _ = search_pattern (*result, new_matrix)
            end = time ()
            sum_time += (end - start)
    OY.append(sum_time)
    
plt.figure(figsize=(20,10))
plt.plot(OX, OY)
plt.show()
```
