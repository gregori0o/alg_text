## Algorytmy wyszukiwania wzorców

#### i. naiwny

```python
def naive_string_matching (text, pattern, arr = None):
    result = []
    for s in range (len(text)-len(pattern)+1):
        if pattern == text[s:s+len(pattern)]:
            result.append(s)
    return result            
```

#### ii. automat skończony

```python
def transition_table (pattern):
    def make_alphabet ():
        alphabet = set ()
        for i in range (len(pattern)):
            alphabet.add(pattern[i])
        return list(alphabet)
    
    def is_sufiks (suf, text):
        for i in range (len (suf) - 1, -1, -1):
            if suf[-i] != text[-i]:
                return False
        return True
    
    result = []
    alphabet = make_alphabet()
    for q in range (len(pattern) + 1):
        result.append({})
        for letter in alphabet:
            k = min(len(pattern)+1, q + 2)
            while True:
                k = k-1
                if is_sufiks(pattern[:k], pattern[:q] + letter):
                    break
            result[q][letter] = k
    return result

def fa_string_matching (text, pattern, delta = None):
    if delta == None: delta = transition_table (pattern)
    result = []
    q = 0
    for s in range (len(text)):
        q = delta [q].get(text[s], 0)
        if (q == len(pattern)):
            result.append(s+1-q)
    return result
```

#### iii. algorytm KMP

```python
def prefix_function (pattern):
    pi = [0]
    k = 0
    for q in range (1, len(pattern)):
        while k > 0 and pattern [k] != pattern [q]:
            k = pi[k-1]
        if pattern[k] == pattern[q]:
            k += 1
        pi.append(k)
    return pi

def kmp_string_matching (text, pattern, pi = None):
    if pi == None: pi = prefix_function (pattern)
    result = []
    q = 0
    for i in range(0, len(text)):
        while q > 0 and pattern [q] != text[i]:
            q = pi[q-1]
        if pattern[q] == text [i]:
            q += 1
        if q == len(pattern):
            result.append (i + 1 - q)
            q = pi[q-1]
    return result
```

## Pomiar czasu

Pomiar czasu odbywa się za pomocą funkcji 'measure_time'. Funkcja 'measure_time_of_all' pozwala policzyć jednocześnie czas preprocessingu oraz właściwego poszukiwania.

```python
import time

def measure_time (function, pattern, text = None, arr = None):
    tim = 0
    result = []
    if text == None:
        start = time.time ()
        result = function (pattern)
        end = time.time ()
        tim = end - start
    else:
        start = time.time ()
        result = function (text, pattern, arr)
        end = time.time ()
        tim = end - start
    return tim, result

def measure_time_of_all (preprocessing, process, text, pattern):
    preprocessing_time = 0
    processing_time = 0
    if preprocessing != None:
        preprocessing_time, prep = measure_time (preprocessing, pattern)
        processing_time, _ = measure_time (process, pattern, text, prep)
    else:
        processing_time, _ = measure_time (process, pattern, text)
    return preprocessing_time, processing_time, preprocessing_time + processing_time
```

#### Testy pomiaru czasu

```python
from random import randint
import matplotlib.pyplot as plt
import string

def generator_text (N, alphabet, pattern = None, k = 0):
    size = len(alphabet)
    text = ""
    if k > 0:
        N -= k * len(pattern)
    if N <= 0:
        N = 1
    for _ in range (N):
        t = randint (0, size-1)
        text += alphabet [t]
    for _ in range (k):
        t = randint (0, len(text)-1)
        text = text[:t] + pattern + text [t:]
    return text
```

Generator tekstu generuje łańcuch znaków długości N zbudowany z małych liter alfabetu bez polskich znaków. Ma dodatkową funkcję w postaci wpisania w losowe miejsca zadanego wzorca.

Testy zostały podzielone na trzy części. Najpierw testuję dla wzorca długości 10 i tekstu długości od 5000 do 100000 z krokiem 2500 czas działania całego algorytmu. Następnie dla stałego wzorca długości 10 mierzę czas bez preprocessingu. Ostatni test dotyczy samego preprocessingu dla wzorców długości od 1 do 50 z krokiem 1. Wyniki zostały przedstawione na wykresach. Na zielono algorytm naiwny, czerwony to automat skończony, natomiast na niebiesko KMP.

```python
def test_length_of_text ():
    alphabet = string.ascii_lowercase
    naive_time = []
    fa_time = []
    kmp_time = []
    text_len = []
    for leng in range (5000, 100001, 2500):
        pattern = generator_text (10, alphabet)
        text_len.append(leng)
        text = generator_text (leng, alphabet, pattern, leng // 500)
        naive, _ = measure_time(naive_string_matching, pattern, text)
        fa, _ = measure_time(fa_string_matching, pattern, text)
        kmp, _ = measure_time(kmp_string_matching, pattern, text)
        naive_time.append(naive)
        fa_time.append(fa)
        kmp_time.append(kmp)
    plt.figure(figsize=(15,10))
    plt.plot (text_len, naive_time, '-g')
    plt.plot (text_len, fa_time, '-r')
    plt.plot (text_len, kmp_time, '-b')

test_length_of_text ()
```

Algorytm KMP jest najszybszy i wyróżnia się na tle pozostałych. Czasy pozostałych dwóch algorytmów nie różnią się bardzo od siebie. Wszystkie algorytmy uzyskują wykres, który możemy przyblizyć wykresem liniowym. Algorytm naiwny radzi sobie całkiem dobrze, wynika to dużego alfabetu co powoduje rzadkie wystąpienie wzorca, możemy przypuszcać, zę pojawia się on tylko w miejscach gdzie został doklejony w funkcji generator_text.

```python
def test_without_preprocessing ():
    alphabet = string.ascii_lowercase
    pattern = generator_text (10, alphabet)
    delta = transition_table (pattern)
    pi = prefix_function (pattern)
    naive_time = []
    fa_time = []
    kmp_time = []
    text_len = []
    for leng in range (5000, 100001, 2500):
        text_len.append(leng)
        text = generator_text (leng, alphabet, pattern, leng // 500)
        naive, _ = measure_time(naive_string_matching, pattern, text)
        fa, _ = measure_time(fa_string_matching, pattern, text, delta)
        kmp, _ = measure_time(kmp_string_matching, pattern, text, pi)
        naive_time.append(naive)
        fa_time.append(fa)
        kmp_time.append(kmp)
    plt.figure(figsize=(15,10))
    plt.plot (text_len, naive_time, '-g')
    plt.plot (text_len, fa_time, '-r')
    plt.plot (text_len, kmp_time, '-b')

test_without_preprocessing ()
```

Możemy zaobserwować zależność czasów w następującej postaci: naiwny > automat skończony > KMP.

```python
def test_preprocessing ():
    alphabet = string.ascii_lowercase
    fa_time = []
    kmp_time = []
    text_len = []
    for leng in range (1, 51, 1):
        text_len.append(leng)
        pattern = generator_text (leng, alphabet)
        fa, _= measure_time(transition_table, pattern)
        kmp, _ = measure_time(prefix_function, pattern)
        fa_time.append(fa)
        kmp_time.append(kmp)
    plt.figure(figsize=(15,10))
    plt.plot (text_len, fa_time, '-r')
    plt.plot (text_len, kmp_time, '-b')

test_preprocessing ()
```

Czas preprocessingu dla algorytmu KMP jest bardzo mały w porównaniu do czasu dla automatu skończonego. Czas preprocessingu dla automatu skończonego szybko rośnie.

Algorytm KMP jest najbardziej optymalny, preprocessing tego algorytmu rośnie liniowo do długości wzorca oraz ma liniowy algorytm wyszukiwania wzorca. Algorytm automatu skończonego jest dobry tylko dla krótkich wzorców, wtedy czasy działania są nieznacznie większe od KMP. Zaletą algorytmu naiwnego jest bardzo prosta implementacja. Przy wyszukiwaniu wyrażenia w tekście naturalnym, gdzie wzorzec występuje rzadko oraz rzadko wystepują prefiksy wzorca działa on szybko, jednak nie nadaje się do zastosowania wszędzie co udowodnię poniżej pokazując jak wolno może działać w porównaniu do pozostałych.

## Specjalne teksty i wzorce

Przedstawię przykłądy pokazujące różnice w czasie działania algorytmów. Dla każdego algorytmu wykonam 20 prób i policzę iloraz czasów.

Wzorzec jest ciągiem 40 000 liter a. Tekst składa się z 10 wzorców poprzeplatanych literą "b". Dzięki temu iloraz czasu działania algorytmu naiwnego do algorytmu automatu skończonego jest większy niż 10 oraz do algorytmu KMP jest większy niż 5. Pomiary dokonywane bez preprocessingu.

```python
text = ("a"*40000 + "b")*9 + "a"*40000
pattern = "a"*40000
delta = transition_table (pattern)
pi = prefix_function (pattern)
n = 0
f = 0
k = 0
for _ in range (10):
    naive, _ = measure_time(naive_string_matching, pattern, text)
    fa, _ = measure_time(fa_string_matching, pattern, text, delta)
    kmp, _ = measure_time(kmp_string_matching, pattern, text, pi)
    n += naive
    f += fa
    k += kmp
    
print ("Czas wyszukiwania dla algorytmu naiwnego wynosi", n/10)
print ("Czas wyszukiwania dla automatu skończonego wynosi", f/10)
print ("Czas wyszukiwania dla algorytmu KMP wynosi", k/10)
print ("Czas dla naiwnego jest", n/f, "raza wolniejszy niż dla automatu.")
print ("Czas dla naiwnego jest", n/k, "raza wolniejszy niż dla KMP.")
```

Wzorzec jest listą wszystkich małych liter w tablicy ASCII. Obliczam czasy processingu i dla takiego wzorca iloraz czasu preprocessingu automatu do preprocessingu KMP jest ponad 1000 krotnie większy. Wynika to z faktu, że preprocessing KMP ma złożoność O(m) gdzie m jest długością wzorca. Natomiast dla automatu ma złożoność O(m^3 * |alphabet|) co dla takiego wzorca sprowadza się do O(m^4) z czego wynika różnica w czasie.

```python
pattern = string.ascii_lowercase
f = 0
k = 0
for _ in range (30):
    fa, _= measure_time(transition_table, pattern)
    kmp, _= measure_time(prefix_function, pattern)
    f += fa
    k += kmp
print ("Czas preprocessingu dla automatu skończonego wynosi", f/30)
print ("Czas preprocessingu dla algorytmu KMP wynosi", k/30)
print ("Czas dla automatu jest", f/k, "raza wolniejszy niż dla KMP.")
```
